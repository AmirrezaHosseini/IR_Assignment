```python
content=df["content"]
tokenizer=WordTokenizer()
normalizer = Normalizer()

finall_output=[]
for i in range(len(content)):
  try:
      nor=normalizer.normalize(content[i])
      tokens=tokenizer.tokenize(nor)
      finall_output.append(tokens)
      df.at[i,'Normalized_output'] = tokens
  except:
    finall_output.append(content[i])
    df.at[i,'Normalized_output'] = content[i]
    pass

df.head()
```

```python
df = pd.read_csv("/content/MovieSummaries/plot_summaries.txt", delimiter = "\t",names=["id","text"])
df.info()
```

```python
df.head()
```

```python
import re

text = df["text"]
normalizedRow =[]
for value in text:
    normalizedRow.append(re.sub(" \W","",value))
    #result = tokenizer.tokenize(normalizedRow)
    #print(normalizedRow)
```

```python
normalizedRow[0]
```

```python
import nltk
nltk.download('all')
```

```python
from nltk.stem.porter import PorterStemmer
from nltk import word_tokenize
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
output_stemmed=[]
for text in normalizedRow:
    words = word_tokenize(text)
    filtered_words = [word for word in words if word not in stop_words]
    porter = PorterStemmer()
    output_stemmed.append([porter.stem(word) for word in filtered_words])
output_stemmed[0]
```

```python
df["Normal Text"]=output_stemmed
df.head()
```

```python
df.to_csv("MovieSummaries.csv")
```
